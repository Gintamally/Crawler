```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
url = 'https://www.ptt.cc/bbs/NBA/index.html'

r = requests.get(url)
```

```python
soup = BeautifulSoup(r.text, 'lxml')
print(soup)
```

```python
titles = []
for t in soup.find_all(class_='title'):
    titles.append(t.text.strip())

authors = []
for a in soup.find_all(class_='author'):
    authors.append(a.text)

dates =[]
for d in soup.find_all(class_='date'):
    dates.append(d.text)
    
zipdata=list(zip(authors,titles,dates))
```

```python
df = pd.DataFrame(zipdata, columns=['Author','Title','Date'])
df
```

###  function

```python
def parse(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text,'lxml')
    
    titles = []
    for t in soup.find_all(class_='title'):
        titles.append(t.text.strip())

    authors = []
    for a in soup.find_all(class_='author'):
        authors.append(a.text)

    dates =[]
    for d in soup.find_all(class_='date'):
        dates.append(d.text)
    
    zipdata=list(zip(authors,titles,dates))

    df = pd.DataFrame(zipdata, columns=['Author','Title','Date'])
    return(df)
```

```python
parse('https://www.ptt.cc/bbs/movie/index.html')
```
